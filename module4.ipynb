```python
import pandas as pd

# Load the schools_complete file into a data frame
schools_df = pd.read_csv('C:/Users/CDiFa/OneDrive/Uni - Data Analytics/Module 4/Attempt 3/PyCitySchools/Resources/schools_complete.csv')

# Load the students_complete file into a data frame
students_df = pd.read_csv('C:/Users/CDiFa/OneDrive/Uni - Data Analytics/Module 4/Attempt 3/PyCitySchools/Resources/students_complete.csv')
```

```python
# Calculate the total number of unique schools
total_schools = schools_df['School ID'].nunique()

# Calculate the total number of students
total_students = students_df['Student ID'].nunique()

# Calculate the total budget
total_budget = schools_df['budget'].sum()

# Calculate the average maths score
avg_math_score = students_df['maths_score'].mean()

# Calculate the average reading score
avg_reading_score = students_df['reading_score'].mean()

# Calculate the percentage of students who passed maths
students_df['pass_math'] = students_df['maths_score'] >= 70
perc_pass_math = students_df['pass_math'].mean() * 100

# Calculate the percentage of students who passed reading
students_df['pass_reading'] = students_df['reading_score'] >= 70
perc_pass_reading = students_df['pass_reading'].mean() * 100

# Calculate the percentage of students who passed both maths and reading
students_df['pass_both'] = (students_df['maths_score'] >= 70) & (students_df['reading_score'] >= 70)
perc_pass_both = students_df['pass_both'].mean() * 100

# Create a new dataframe for the above calculations
area_summary = pd.DataFrame({
    'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [avg_math_score],
    'Average Reading Score': [avg_reading_score],
    '% Passing Math': [perc_pass_math],
    '% Passing Reading': [perc_pass_reading],
    '% Overall Passing': [perc_pass_both]
})

display(area_summary)
```

```python
# Calculate the school type
per_school_types = schools_df.set_index(['school_name'])['type']

# Calculate the total student count
per_school_counts = students_df['school_name'].value_counts()

# Calculate the per capita spending
per_school_budget = schools_df.set_index(['school_name'])['budget']
per_school_capita = per_school_budget / per_school_counts

# Calculate the average test scores
per_school_math = students_df.groupby(['school_name'])['maths_score'].mean()
per_school_reading = students_df.groupby(['school_name'])['reading_score'].mean()

# Calculate the number of schools with maths scores of 50 or higher
passing_math = students_df[students_df['maths_score'] >= 50]
per_school_passing_math = passing_math.groupby(['school_name'])['Student ID'].count() / per_school_counts * 100

# Calculate the number of schools with reading scores of 50 or higher
passing_reading = students_df[students_df['reading_score'] >= 50]
per_school_passing_reading = passing_reading.groupby(['school_name'])['Student ID'].count() / per_school_counts * 100

# Calculate the schools that passed both maths and reading with scores of 50 or higher
overall_passing = students_df[(students_df['maths_score'] >= 50) & (students_df['reading_score'] >= 50)]
per_school_passing_overall = overall_passing.groupby(['school_name'])['Student ID'].count() / per_school_counts * 100

# Create a new dataframe for the above calculations
per_school_summary = pd.DataFrame({
    'School Type': per_school_types,
    'Total Students': per_school_counts,
    'Total School Budget': per_school_budget,
    'Per Student Budget': per_school_capita,
    'Average Math Score': per_school_math,
    'Average Reading Score': per_school_reading,
    '% Passing Math': per_school_passing_math,
    '% Passing Reading': per_school_passing_reading,
    '% Overall Passing': per_school_passing_overall
})

display(per_school_summary)
```

```python
# sort the dataframe by % Overall Passing in descending order
top_performing_schools = per_school_summary.sort_values('% Overall Passing', ascending=False)

# select the top 5 schools
top_performing_schools = top_performing_schools.head(5)

# display the top performing schools
print(top_performing_schools)
```

```python
# sort the dataframe by % Overall Passing in ascending order
bottom_performing_schools = per_school_summary.sort_values('% Overall Passing', ascending=True)

# select the bottom 5 schools
bottom_performing_schools = bottom_performing_schools.head(5)

# display the bottom performing schools
print(bottom_performing_schools)
```

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
# Categorize the spending based on the bins
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=group_names)
spending_math_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = per_school_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = per_school_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing = per_school_summary.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
# Assemble into data frame
spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing": overall_passing})
# Minor data munging
spending_summary = spending_summary[["Average Math Score",
                                     "Average Reading Score",
                                     "% Passing Math", "% Passing Reading",
                                     "% Overall Passing"]]
spending_summary["Average Math Score"]=spending_summary["Average Math Score"].map("{:,.2f}".format)
spending_summary["Average Reading Score"]=spending_summary["Average Reading Score"].map("{:,.2f}".format)
spending_summary["% Passing Math"]=spending_summary["% Passing Math"].map("{:,.2f}".format)
spending_summary["% Passing Reading"]=spending_summary["% Passing Reading"].map("{:,.2f}".format)
spending_summary["% Overall Passing"]=spending_summary["% Overall Passing"].map("{:,.2f}".format)
# Display results
spending_summary
```
